```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    'Model': [
        'Logistic Regression (cuML)',
        'Random Forest (cuML)',
        'XGBoost',
        'Plain Federated Learning',
        'Encrypted Federated Learning',
        'Encrypted FL (Fake Data)',
        'Neural Network'
    ],
    'Accuracy': [0.8275, 0.9120, 0.88, 0.3563, 0.9039, 0.8630, 0.85],
    'Precision': [0.6801, 0.8820, 0.89, 0.3115, 0.8383, 0.8620, 0.85],
    'Recall': [0.8022, 0.8200, 0.87, 0.9560, 0.9417, 0.8915, 0.81],
    'F1-Score': [0.7362, 0.8500, 0.88, 0.4699, 0.8618, 0.8628, 0.82],
    'ROC-AUC': [0.8801, 0.8563, 0.8832, None, None, None, None],
}

df = pd.DataFrame(data)

# ===================
# COMPARISON CHARTS
# ===================

metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
colors = ['#4285F4', '#EA4335', '#FBBC05', '#34A853']

x = np.arange(len(df['Model']))
width = 0.18

plt.figure(figsize=(14, 8))
for i, metric in enumerate(metrics):
    plt.bar(x + i*width, df[metric], width, label=metric, color=colors[i])

plt.xticks(x + width*1.5, df['Model'], rotation=45, ha='right')
plt.title('Model Performance Comparison', fontsize=16, fontweight='bold')
plt.ylabel('Score')
plt.legend()
plt.tight_layout()
plt.show()

# ====================
# ROC-AUC COMPARISON
# ====================
roc_df = df.dropna(subset=['ROC-AUC'])

plt.figure(figsize=(8, 5))
plt.bar(roc_df['Model'], roc_df['ROC-AUC'], color='#6A5ACD')
plt.title('ROC-AUC Comparison', fontsize=14, fontweight='bold')
plt.ylabel('ROC-AUC')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```
